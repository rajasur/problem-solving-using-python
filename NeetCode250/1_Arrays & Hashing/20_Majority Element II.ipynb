# 229. Majority Element II

**Difficulty:** Medium

## Problem Statement

Given an integer array of size `n`, find all elements that appear more than `⌊n / 3⌋` times.

---

## Examples

### Example 1

**Input:**
```text
nums = [3,2,3]
```

**Output:**
```text
[3]
```

---

### Example 2

**Input:**
```text
nums = [1]
```

**Output:**
```text
[1]
```

---

### Example 3

**Input:**
```text
nums = [1,2]
```

**Output:**
```text
[1,2]
```

---

## Constraints

- `1 <= nums.length <= 5 × 10^4`
- `-10^9 <= nums[i] <= 10^9`

---

## Follow-up

Can you solve the problem in **O(n)** time and **O(1)** extra space?

````python
"""
Optimal Solution
TC -> O(n)
SC -> O(1)
"""
from typing import List
import math
class Solution:
    def majorityElement(self, nums: List[int]) -> List[int]:
        major_ele_1, major_ele_2 = None, None 
        count_1, count_2 = 0, 0
        n = len(nums)
        result = []

        # Boyer-Moore Voting
        for idx in range(n):
            if major_ele_1 == nums[idx]:
                count_1 += 1 
            elif major_ele_2 == nums[idx]:
                count_2 += 1 
            elif count_1 == 0:
                major_ele_1 = nums[idx]
                count_1 = 1 
            elif count_2 == 0:
                major_ele_2 = nums[idx]
                count_2 = 1 
            else:
                count_1 -= 1
                count_2 -= 1 
        
        # Verification
        if nums.count(major_ele_1) > math.floor(n//3):
            result.append(major_ele_1)
        if nums.count(major_ele_2) > math.floor(n//3):
            result.append(major_ele_2)
        return result 
    
if __name__ == "__main__":
    nums = [3, 2, 3]
    sol = Solution()
    print(sol.majorityElement(nums))
````
