# 347. Top K Frequent Elements

**Difficulty:** Medium

## Problem Statement

Given an integer array `nums` and an integer `k`, return the `k` most frequent elements.

You may return the answer in **any order**.

---

## Examples

### Example 1

**Input:**
```text
nums = [1,1,1,2,2,3], k = 2
```

**Output:**
```text
[1,2]
```

---

### Example 2

**Input:**
```text
nums = [1], k = 1
```

**Output:**
```text
[1]
```

---

### Example 3

**Input:**
```text
nums = [1,2,1,2,1,2,3,1,3,2], k = 2
```

**Output:**
```text
[1,2]
```

---

## Constraints

- `1 <= nums.length <= 10^5`
- `-10^4 <= nums[i] <= 10^4`
- `1 <= k <=` the number of unique elements in `nums`
- It is guaranteed that the answer is unique.

---

## Follow-up

Can you design an algorithm with a time complexity **better than** `O(n log n)`?

````python
"""
Optimal Solution
Time Complexity: O(n)
Space Complexity: O(n)
"""

from typing import List
from collections import Counter
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = Counter(nums)
        
        #Create Empty Frequency Bucket
        bucket = [[] for _ in range(len(nums) + 1)]

        # Implement Bucket sort algorithm
        for number, count in count.items():
            bucket[count].append(number)

        result = []
        for frequency in range(len(bucket)-1, 0 , -1):
            result.extend(bucket[frequency])
            if len(result) == k:
                return result

if __name__ == "__main__":
    sol = Solution()
    arr = [1,2,1,2,1,2,3,1,3,2]
    k = 2
    print(sol.topKFrequent(arr, k))
````
