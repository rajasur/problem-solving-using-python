# 169. Majority Element

**Difficulty:** Easy

## Problem Statement

Given an array `nums` of size `n`, return the **majority element**.

The majority element is the element that appears **more than** `⌊n / 2⌋` times. You may assume that the majority element always exists in the array.

---

## Examples

### Example 1

**Input:**
```text
nums = [3,2,3]
```

**Output:**
```text
3
```

---

### Example 2

**Input:**
```text
nums = [2,2,1,1,1,2,2]
```

**Output:**
```text
2
```

---

## Constraints

- `n == nums.length`
- `1 <= n <= 5 × 10^4`
- `-10^9 <= nums[i] <= 10^9`
- The input is generated such that a majority element will always exist.

---

## Follow-up

Can you solve the problem in **O(n)** time and **O(1)** extra space?

````python
"""
Optimal Solution
TC -> O(n)
SC -> O(1)
"""
from typing import List
import math
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        majority_elemnt = None
        count = 0
        n = len(nums) 

        # Boyer-Moore Voting
        for idx in range(n):
            if majority_elemnt == nums[idx]:
                count += 1 
            elif count == 0:
                majority_elemnt = nums[idx]
                count = 1 
            else:
                count -= 1 

        # Verification
        if nums.count(majority_elemnt) > math.floor(n//2):
            return majority_elemnt 
        else:
            return -1

if __name__ == "__main__":
     nums = [2,2,1,1,1,2,2] 
     sol = Solution()
     print(sol.majorityElement(nums))
````
