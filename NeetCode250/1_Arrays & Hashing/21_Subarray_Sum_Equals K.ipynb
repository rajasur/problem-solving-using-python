# 560. Subarray Sum Equals K.

**Difficulty:** Medium

## Problem Statement

Given an array of integers `nums` and an integer `k`, return the total number of **subarrays** whose sum equals to `k`.

A **subarray** is a contiguous non-empty sequence of elements within an array.

---

## Examples

### Example 1

**Input:**
```text
nums = [1,1,1], k = 2
```

**Output:**
```text
2
```

**Explanation:**

The subarrays whose sum is `2` are:

- `[1,1]` (indices `0-1`)
- `[1,1]` (indices `1-2`)

---

### Example 2

**Input:**
```text
nums = [1,2,3], k = 3
```

**Output:**
```text
2
```

**Explanation:**

The subarrays whose sum is `3` are:

- `[1,2]`
- `[3]`

---

## Constraints

- `1 <= nums.length <= 2 × 10^4`
- `-1000 <= nums[i] <= 1000`
- `-10^7 <= k <= 10^7`

````python
"""
TC -> O(n)
SC -> O(n)
"""
from typing import List
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        n = len(nums)
        prefix_sum = [0] * n
        count = 0 
        
        # Build prefix sum array
        prefix_sum[0] = nums[0]
        for idx in range(1, n):
            prefix_sum[idx] = prefix_sum[idx - 1] + nums[idx]

        # HashMap: {prefix_sum: frequency}
        freq_sum = {}
        for idx in range(n):
            # Subarray starting from index 0
            if prefix_sum[idx] == k:
                count += 1

            # Check if (current_prefix_sum - k) exists
            val = prefix_sum[idx] - k
            if val in freq_sum:
                count += freq_sum[val]

            # Update frequency of current prefix sum 
            if prefix_sum[idx] not in freq_sum:
                freq_sum[prefix_sum[idx]] = 0 
            freq_sum[prefix_sum[idx]] += 1 
        return count 
    
if __name__ == "__main__":
    nums = [1,2,3]
    k = 3
    sol = Solution()
    print(sol.subarraySum(nums, k))
````
