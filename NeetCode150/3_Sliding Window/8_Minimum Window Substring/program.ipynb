# 76. Minimum Window Substring

**Solved**  
**Difficulty:** Hard  

---

## ðŸ§© Problem Description

Given two strings `s` and `t` of lengths `m` and `n` respectively, return the **minimum window substring** of `s` such that **every character in `t` (including duplicates)** is included in the window.

- If there is **no such substring**, return an empty string `""`.
- The test cases are generated such that the answer is **unique**.

---

## ðŸ“˜ Examples

### Example 1
**Input:**
```text
s = "ADOBECODEBANC", t = "ABC"
````

**Output:**

```text
"BANC"
```

**Explanation:**
The minimum window substring `"BANC"` includes `'A'`, `'B'`, and `'C'` from string `t`.

---

### Example 2

**Input:**

```text
s = "a", t = "a"
```

**Output:**

```text
"a"
```

**Explanation:**
The entire string `s` is the minimum window.

---

### Example 3

**Input:**

```text
s = "a", t = "aa"
```

**Output:**

```text
""
```

**Explanation:**
Both `'a'` characters from `t` must be included in the window.
Since the largest possible window in `s` contains only one `'a'`, return an empty string.

---

## ðŸ”’ Constraints

* `m == s.length`
* `n == t.length`
* `1 <= m, n <= 10^5`
* `s` and `t` consist of **uppercase and lowercase English letters**

---

## ðŸš€ Follow-up

Can you find an algorithm that runs in **O(m + n)** time?

`````python
"""
Time & Space Complexity
Time complexity: O(n)
Space complexity: O(1)
"""
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        hash = [0] * 256
        left, right = 0, 0
        count = 0 
        min_len = float("inf")
        start_index = -1

        for i in range(len(t)):
            hash[ord(t[i])] += 1

        while right < len(s):
            if hash[ord(s[right])] > 0:
                count += 1
            hash[ord(s[right])] -= 1
            right += 1

            while count == len(t):
                if right - left < min_len:
                    min_len = right - left
                    start_index = left

                hash[ord(s[left])] += 1
                if hash[ord(s[left])] > 0:
                    count -= 1
                left += 1

        if min_len == float("inf"):
            return ""
        return s[start_index:start_index + min_len]


if __name__ == "__main__":
    s = "ADOBECODEBANC"
    t = "ABC"
    solution = Solution()
    result = solution.minWindow(s, t)
    print(result)  # Output: "BANC"
`````
