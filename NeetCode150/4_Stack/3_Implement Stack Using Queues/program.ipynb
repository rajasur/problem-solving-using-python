```python
class MyStack:

    def __init__(self):
        self.queue = []
        

    def push(self, x: int) -> None:
        self.queue.append(x)
        # Rotate the queue to make the last element first
        for i in range(len(self.queue) - 1):
            self.queue.append(self.queue.pop(0))
        

    def pop(self) -> int:
        return self.queue.pop(0)
        

    def top(self) -> int:
        return self.queue[0]
        

    def empty(self) -> bool:
        return len(self.queue) == 0


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()

if __name__ == "__main__":
    stack = MyStack()
    operations = ["MyStack", "push", "push", "top", "pop", "empty"]
    values = [[], [1], [2], [], [], []]
    output = []
    
    for i, op in enumerate(operations):
        if op == "MyStack":
            output.append(None)
        elif op == "push":
            stack.push(values[i][0])
            output.append(None)
        elif op == "top":
            output.append(stack.top())
        elif op == "pop":
            output.append(stack.pop())
        elif op == "empty":
            output.append(stack.empty())
    
    print("Output:", output)
    
```
